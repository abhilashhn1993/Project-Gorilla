```python
df['resale_value'].describe()

bins = ['Less','Good','Higher']
df['resale_value'] = pd.cut(x=df['resale_value'], bins=[0,12,20,100], labels=bins)
df.head()
```

```python
#df['Length'].describe()

types = ['Compact', 'Sedan', 'SUV']
df['Type'] = pd.cut(x=df['Width'], bins=[0,185,215,225], labels=types)
df.head()
```
