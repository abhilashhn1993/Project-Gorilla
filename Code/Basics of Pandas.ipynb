# Exploratory Data Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Load & read the data 

```python
df = pd.read_csv('Car_sales.csv')
#df_x = pd.read_excel('data.xls')
```

```python
df.head()
```

### The first Look

```python
df.info()
```

```python
df.columns
```

```python
df = df.rename(columns={'__year_resale_value': 'resale_value', 'Price_in_thousands': 'Price', 'Vehicle_type': 'Class'})
```

```python
df.head()
```

```python
df = df.drop_duplicates(keep='first')
```

```python
df = df.drop(['Latest_Launch'], axis=1)
```

```python
df.shape
```

```python
df = df.dropna()
df.isna().sum()
```

### Data Subsetting

```python
#df[['Manufacturer','Model','Price','resale_value']][5:15]
#df.iloc[0:5,0:3]

df.loc[df['Manufacturer']=='Toyota'].Class
```

```python
df[(df['Manufacturer']=='Audi') & (df['Horsepower'] > 150)]
```

```python
for index, row in df.iterrows():
    print(index, row[['Manufacturer', 'Model', 'Price']])
```

```python
df.sort_values(['Manufacturer','Price'], ascending=[1,0])
```

```python
df.sort_values(['Manufacturer','Price'], ascending=[1,1])
```

```python
df['resale_value'].describe()
```

```python
bins = ['Less', 'Medium', 'Higher']
```

```python
df['resale_value'] = pd.cut(x=df['resale_value'], bins=[0,12,20,70], labels=bins)
```

```python
df.head()
```

```python
df.resale_value = pd.Categorical(df.resale_value, ['Less','Medium','Higher'], ordered=True)
```

```python
df['Length'].describe()
```

```python
bins = ['Compact','Sedan','SUV']
```

```python
df['Type']=pd.cut(x=df['Length'], bins=[0,185,215,225], labels=bins)
```

```python
df['Type'].value_counts()
```

```python
df['Car_Name'] = df['Manufacturer'] + " " + df['Model']
df.head()
```

```python
df.to_csv('Car_sales_modified.csv')
```
